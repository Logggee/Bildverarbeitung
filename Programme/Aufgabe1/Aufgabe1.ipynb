```python
import numpy as np
import cv2
import math

path = 'Gesicht.jpg'  #Dateipfad des Bildes
src = cv2.imread(path) #Bild einlesen
eyes = list()

#draw point with mouse in the picture
def DrawPoint(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN: #Wenn linke Maustaste gedrückt wurde
        eyes.append((x,y))             #Koordinaten des Punktes speichern
        cv2.circle(src, center=(x,y), radius=0, color=(0,0,255), thickness=10) #roten Kreis an die stelle des Mauszeigers Zeichnen
        
#Berechnen des Winkels zwischen den beiden Punkten        
def CalcAngle(cords):
    deltaX = cords[0][0] - cords[1][0]
    deltaY = cords[0][1] - cords[1][1]
    tan = deltaY/deltaX
    alpha = math.degrees(math.tan(tan))
    return alpha

#Rotieren des Bildes mit berechnetem Winkel
def RotateImage(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result

#Punkt zeichnen
cv2.namedWindow(winname = 'image')
cv2.setMouseCallback('image',DrawPoint)  #Funktion für den Mausinterrupt 

#Output         
while(1):
    cv2.imshow('image',src)  #Ausgabe des Image in einer Unendlichen Schleife
    if len(eyes) == 2:
        angle = CalcAngle(eyes)      #Winkel berechnen
        img = RotateImage(src, angle)  #Rotieren mit berechnetem Winkel
        img = img[10:250, 200:420]     #Zuschneiden des Bildes um Schwarzen Rand wegzuschneiden
        cv2.imshow('image',img)  #Ausgabe des Image in einer Unendlichen Schleife
        break
        
    if cv2.waitKey(20) == ord('q'):  #Abbruch mit der Taste q
        break

print("If you want to safe the picture press s if not press q")
while True:
    if cv2.waitKey(0) == ord('q'):  #Abbruch mit der Taste q
        break
    if cv2.waitKey(0) == ord('s'):  #Speichern des Rotierten Bildes mit Taste s
        cv2.imwrite("RotatetFace.jpg", img) #Image mit diesem Dateinamen speichern
        break
cv2.destroyAllWindows()
```
